### Method - Deleting rows or column

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('train.csv')
```

```python
pd.set_option('display.max_columns',None) # To display all the columns
pd.set_option('display.max_rows',None) # To display all the rows
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum() # to calculate the total numbers of null value in column
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df.isnull()) # in this heatmap the white lines means the null values
```

```python
null_var = df.isnull().sum()/df.shape[0]*100 # calculating the total percentage of null values in all columns
null_var
```

```python
drop_columns = null_var[null_var > 17].keys() # Extracting the columns which have null values greater than 17% so that we can delete the entire column
drop_columns
```

```python
df2_drop_col = df.drop(columns = drop_columns) # The new dataframe after deleting the columns with null values
```

```python
df2_drop_col.shape
```

```python
sns.heatmap(df2_drop_col.isnull())
```

```python
df3_drop_row = df2_drop_col.dropna() # Deleting the entire null where the null value is present
```

```python
df3_drop_row.shape
```

```python
sns.heatmap(df3_drop_row.isnull()) # here the heatmap is clear so there are no any null values
```

```python
df3_drop_row.select_dtypes(include = ['int64','float64']).columns # Extracting all the columns with int and float datatypes
```

```python
sns.distplot(df['MSSubClass'])
sns.distplot(df3_drop_row['MSSubClass'])

# checking the distribution of data with respect to single feature after cleaning with respect to original
# Here both the figures are overlapped , slight difference so clearing the null values doesnot affect so much on this column
```

```python
## Now lets check the distrubition of all the features which has numeric values

num_var = ['MSSubClass', 'LotArea', 'OverallQual', 'OverallCond',
       'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2',
       'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
       'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF',
       'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal',
       'MoSold', 'YrSold', 'SalePrice']

plt.figure(figsize=(25,25))
for i , var in enumerate(num_var):
    plt.subplot(9,4,i+1)  # here (9,4) means there are total 36 columns with numeric values and we want 9 rows and 4 columns
    sns.distplot(df[var],bins = 20)
    sns.distplot(df3_drop_row[var], bins = 20)
```
